```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name','Name_len']].head()
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x < 17 else 'Adult')
```

```python
titanic_df[['Age','Child_Adult']]
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x <17 else('Adult' if x<=60 else 'Elderly'))
```

```python
titanic_df[['Age','Age_cat']]
```

```python
def get_category(age):
    cat = ''
    if age<=10 : cat = 'Baby'
    elif age < 17 : cat = 'Teenager'
    elif age < 60 : cat = 'Adult'
    else : cat = 'Elderly'
    return cat
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
```

```python
titanic_df[['Age','Age_cat']].head(10)
```
